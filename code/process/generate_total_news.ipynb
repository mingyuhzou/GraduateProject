由于数据是按照用户点击行为划分的，想要得到所有的news信息需要先合并

```python
import polars as pl
```

```python
train_path='/home/mingyu/graduationProject/Data/MINDlarge_train/'
valid_path='/home/mingyu/graduationProject/Data/MINDlarge_dev/'
test_valid='/home/mingyu/graduationProject/Data/MINDlarge_test/'
paths=[train_path,valid_path,test_valid]
```

```python
train=pl.read_parquet(train_path+'items.parquet')
valid=pl.read_parquet(valid_path+'items.parquet')
test=pl.read_parquet(test_valid+'items.parquet')
```

```python
drop_cols = ["title_entities", "abstract_entities"]

train = train.drop(drop_cols)
valid = valid.drop(drop_cols)
test  = test.drop(drop_cols)

news = (
    pl.concat([train, valid, test], how="vertical")
      .unique(subset=["news_id"])
)
```

```python
news.to_pandas().info()
```

```python
news.write_parquet('../../Data/news.parquet')
```
