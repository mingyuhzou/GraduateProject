```python
# 请确保您已将 API Key 存储在环境变量 ARK_API_KEY 中
# 初始化Ark客户端，从环境变量中读取您的API Key
client = OpenAI(
    # 此为默认路径，您可根据业务所在地域进行配置
    base_url="https://ark.cn-beijing.volces.com/api/v3",
    # 从环境变量中获取您的 API Key。此为默认方式，您可根据需要进行修改
    api_key=os.environ.get("ARK_API_KEY"),
)
```

```python
imagesResponse = client.images.generate(
    model="doubao-seedream-4-0-250828",
    prompt='''
    生成新闻封面，新闻属性如下：
        类别：lifestyle
        子类别：lifestyleroyals
        标题：The Brands Queen Elizabeth, Prince Charles, and Prince Philip Swear By
        摘要：Shop the notebooks, jackets, and more that the royals can't live without.
    ''',
    size="2K",
    response_format="b64_json",  # 改为 b64_json
    extra_body={"watermark": True},
)

import base64

image_b64 = imagesResponse.data[0].b64_json
save_path = "news_cover.jpg"

with open(save_path, "wb") as f:
    f.write(base64.b64decode(image_b64))

print(f"图片已保存到 {save_path}")
```
